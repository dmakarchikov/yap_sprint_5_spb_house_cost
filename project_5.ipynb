#### Атрибут *floors_total* - количество этажей в доме

**Построим гистаграмму и изучим данные**

```python
data['floors_total'].plot(kind='hist');
```

В Петербурге бóльшая часть жилого фонда - малоэтажные дома. Учитывая, что данные включают область, где высоток еще меньше, можно предположить, что гистограмма похожа на правду.

```python
print(f'Медианное количество этажей в доме: {int(data["floors_total"].mean())} эт.')
```

Медианное количество этажей - 10. Но если подставить его по всем отсутствующим значениям, то появятся записи с квартирами, находящимися на этажах выше высотности дома.  
Чтобы этого избежать сохраним в переменную `floors_dict` словарь с медианной высотностью в зависимости от этажа квартиры - индекса.

```python
floors_dict = (data
               .pivot_table(index='floor',
                            values='floors_total',
                            aggfunc='median'
                            )
               .to_dict('dict')  # Преобразуем сводную таблицу в словарь
               .get('floors_total')  # Нужный нам словарь находится в ключе с именем floors_total
               )
floors_dict
```

**Заполним пустые значения этажности дома**

```python
data['floors_total'] = data['floors_total'].fillna(data['floor'].map(floors_dict))
```

**Убедимся, что все атрибуты заполнились**

```python
print(f'Незаполненных атрибутов floors_total: {data["floors_total"].isna().sum()}')
```

### Поиск дубликатов

**Проверим данные на наличие явных дубликатов**

```python
print(f'Найдено {data.duplicated().sum()} дубликатов')
```

Явных дубликатов не найдено. Переходим к поиску неявных.
Неявные дубликаты часто встречаются в текстовых атрибутах. В нашем датасете есть 1 такой атрибут - `locality_name`.  
Причина дубликатов связана с человеческим фактором:
- пользователь может допустить опечатку
- использование буквы "ё"
- несоблюдение регистров при написании  
  
Также причины могут быть техническими:
- неправильная кодировка текста
- населенный пункт может со временем изменить тип - например, поселок может стать поселком городского типа

**Посмотрим на качество заполнения населенных пунктов**  

```python
print(f'Всего нас. пунтов: {len(data["locality_name"].unique())}')
data['locality_name'].unique()
```

На глаз сложно найти дубликаты, но сразу можно выделить:
- буквы "е" и "ё"
- встречаются населенные пункты с обозначением типа и без  
  
**Создадим новую колонку для обработанного наименования нас. пункта и заменим буквы "ё" на "е"**

```python
data.insert(loc=14,
            column='locality_clean_name',
            value=data['locality_name'].str.lower().str.replace('ё', 'е')
           )
data.head()
```

**Изучим оставшиеся записи**

```python
cities = list(np.sort(data['locality_clean_name'].unique()))
for city in cities:
    print(city)
```

В рамках анализа можно избавиться от типа населенного пункта и оставить только имя собственное.

```python
data['locality_clean_name'] = (
    data['locality_clean_name']
    .str.replace('городской поселок ', '')
    .str.replace('поселок городского типа ', '')
    .str.replace('поселок при железнодорожной станции ', '')
    .str.replace('поселок станции ', '')
    .str.replace('коттеджный поселок ', '')
    .str.replace('поселок ', '')
    .str.replace('садоводческое некоммерческое товарищество ', '')
    .str.replace('садовое товарищество ', '')
    .str.replace('деревня ', '')
    .str.replace('село ', '')
    .str.title()  # сделаем первые буквы заглавными
    )
data.head(15)
```
