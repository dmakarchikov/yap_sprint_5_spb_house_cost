#### Атрибут *ceiling_height* -  высота потолка

Из матрицы корреляции видно, что значение высоты потолков не связано ни с одним другим параметром квартиры. Значение корреляции по всем параметрам меньше 10%. Можно было бы предположить, что высота потолков может быть связана с годом постройки дома, но эта информация не была предоставлена.
Исходя из этого заполним пропуски медианным значением.

```python
data['ceiling_height'].plot(kind='hist', range=(2.4, 4));
```

```python
ceiling_median = round(data['ceiling_height'].median(), 2)
data['ceiling_height'] = data['ceiling_height'].fillna(ceiling_median)
print(f'Медианная высота потолков: {ceiling_median} м.')
```

```python
data['ceiling_height'].plot(kind='hist', range=(2.4, 4));
```

*Вывод:* произвели замену пропусков на медианное значение. Распределение приняло вид распределения Пуассона.

#### Атрибут *floors_total* - количество этажей в доме

**Построим гистаграмму и изучим данные**

```python
data['floors_total'].plot(kind='hist');
```

В Петербурге бóльшая часть жилого фонда - малоэтажные дома. Учитывая, что данные включают область, где высоток еще меньше, можно предположить, что гистограмма похожа на правду.

```python
print(f'Медианное количество этажей в доме: {int(data["floors_total"].mean())} эт.')
```

Медианное количество этажей - 10. Но если подставить его по всем отсутствующим значениям, то появятся записи с квартирами, находящимися на этажах выше высотности дома.  
Чтобы этого избежать сохраним в переменную `floors_dict` словарь с медианной высотностью в зависимости от этажа квартиры - индекса.

```python
floors_dict = (data
               .pivot_table(index='floor',
                            values='floors_total',
                            aggfunc='median'
                            )
               .to_dict('dict')  # Преобразуем сводную таблицу в словарь
               .get('floors_total')  # Нужный нам словарь находится в ключе с именем floors_total
               )
floors_dict
```

**Заполним пустые значения этажности дома**

```python
data['floors_total'] = data['floors_total'].fillna(data['floor'].map(floors_dict))
```

**Убедимся, что все атрибуты заполнились**

```python
print(f'Незаполненных атрибутов floors_total: {data["floors_total"].isna().sum()}')
```

*Вывод:* атрибут `floors_total` важен для выборки, т.к. есть предположение, что квартиры на первых и последних этажах ложны отличаться по цене от аналогичных квартир на других этажах.  Поэтому мы заменили отсутствующие значения на медианные.

Поэтому не будем заполнять эти данные синтетическими значениями, а оставим их пустыми.

#### Атрибут *is_apartment* - апартаменты

### Поиск дубликатов

**Проверим данные на наличие явных дубликатов**

Явных дубликатов не найдено. Переходим к поиску неявных.
Неявные дубликаты часто встречаются в текстовых атрибутах. В нашем датасете есть 1 такой атрибут - `locality_name`.  
Причина дубликатов связана с человеческим фактором:
- пользователь может допустить опечатку
- использование буквы "ё"
- несоблюдение регистров при написании  
  
Также причины могут быть техническими:
- неправильная кодировка текста
- населенный пункт может со временем изменить тип - например, поселок может стать поселком городского типа

**Посмотрим на качество заполнения населенных пунктов**  

```python
print(f'Всего нас. пунтов: {len(data["locality_name"].unique())}')
data['locality_name'].unique()
```

На глаз сложно найти дубликаты, но сразу можно выделить:
- буквы "е" и "ё"
- встречаются населенные пункты с обозначением типа и без  
  
**Создадим новую колонку для обработанного наименования нас. пункта и заменим буквы "ё" на "е"**

```python
data.insert(loc=14,
            column='locality_clean_name',
            value=data['locality_name'].str.lower().str.replace('ё', 'е')
           )
data.head()
```

**Изучим оставшиеся записи**

```python
cities = list(np.sort(data['locality_clean_name'].unique()))
for city in cities:
    print(city)
```

В рамках анализа можно избавиться от типа населенного пункта и оставить только имя собственное.

```python
data['locality_clean_name'] = (
    data['locality_clean_name']
    .str.replace('городской поселок ', '')
    .str.replace('поселок городского типа ', '')
    .str.replace('поселок при железнодорожной станции ', '')
    .str.replace('поселок станции ', '')
    .str.replace('коттеджный поселок ', '')
    .str.replace('поселок ', '')
    .str.replace('садоводческое некоммерческое товарищество ', '')
    .str.replace('садовое товарищество ', '')
    .str.replace('деревня ', '')
    .str.replace('село ', '')
    .str.title()  # сделаем первые буквы заглавными
    )
data.head(15)
```

```python
print(f'Всего нас. пунтов после обработки дубликатов: {len(data["locality_clean_name"].unique())}')
```

Короткий вывод: мы добавили новый атрибут `locality_clean_name` с очищенными данными по местоположению объекта недвижимости. Количество уникальных населенных пунтков снизилось почти на 20% с 365 до 306.

### Тип этажа квартиры

**Добавим атрибут `floor_type` на основе атрибута `floor`. Нам нужно разделить первые, последние и все остальные этажи.**

```python
def find_floor_type(row):
    '''
    Функция для определения типа этажа.
    Используемые типы:
    - Первый
    - Последний
    - Другой
    '''
    if row['floor'] == 1:
        return 'Первый'
    elif row['floor'] == row['floors_total']:
        return 'Последний'
    return 'Другой'


data.insert(loc=13,
            column='floor_type',
            value=data.apply(find_floor_type, axis=1)
           )
data.head()
```

### Оценка влияния на стоимость квартиры

Можно сделать предположение, что основными факторами, влияющими на стоимость квартиры, являются:
- общая площадь
- жилая площадь
- площадь кухни
- количество комнат
- этаж, на котором расположена квартира

Для проверки предположений посмотрим на коэффициент корреляции между стоимостью квартиры и указанными атрибутами.

```python
# Список атрибутов для построения корреляции
corr_columns = ['last_price',
                'total_area',
                'living_area',
                'kitchen_area',
                'rooms',
                # 'floor_type',  # не является числовым
                'exp_week_day',
                'exp_month',
                'exp_year'
               ]

# Создадим новый датафрейм для матрицы корреляции
price_corr = pd.DataFrame()
'''
Проходим по каждому атрибуту из corr_columns и добавляем корреляцию
по нему к last_price в датафрейм price_corr
'''
for col in corr_columns:
    price_corr.loc['last_price', col] = data.corr().loc['last_price', col]

price_corr
```

2. Чуть меньше, но тоже сильная, сввязь с жилой площадью и площадью кухни. Это объясняется сильной свзяью между общей площадью и жилыми с кухней. Все это видно на графиках рассевания.
3. Чем больше комнат, тем выше стоимость квартиры. Это утвержддение также косвенно связано с тем, что количество комнат прямо связано с общей площадью.

```python
# Группируем данные по году и месяцу
median_price_by_month = (data.query('locality_clean_name == "Санкт-Петербург"')
                         .groupby(['exp_year',
                                   'exp_month']
                                 )['last_price']
                         .median()
                         .reset_index()
                        )
# Добавляем столбец с единицей для дня
median_price_by_month['exp_day'] = 1
# Переименуем столбцы
median_price_by_month.columns = ['year', 'month', 'median_price', 'day']

# Добавим стобец с датой
median_price_by_month['month_year'] = (pd.
                                       to_datetime(
                                           median_price_by_month[['year',
                                                                  'month',
                                                                  'day']]
                                       )
                                       )

# Удаляем вспомогательные колонки
median_price_by_month = (median_price_by_month
                         .drop(columns=['year',
                                        'month',
                                        'day'
                                       ]
                              )
                        )

# Удаляем выброс в первом месяце
median_price_by_month = (median_price_by_month
                         .query('median_price < 7000000')
                        )

# Строим график
median_price_by_month.plot('month_year',
         'median_price',
         linestyle='-',
         marker='o', figsize=(15, 5), grid=True
        );
plt.title('median_price_by_month')
plt.xticks(rotation=90)

plt.show;
```
